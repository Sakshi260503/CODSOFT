```python
data=df.values
X=data[:,0:4]
Y=data[:,4]
X
```

```python
Y
```

```python
#splitting the data into testing and training model
from sklearn.model_selection import train_test_split

X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2)
print(Y_train)
```

```python
from sklearn.svm import SVC
model=SVC()
model.fit(X_train,Y_train)
```

```python
pred=model.predict(X_test)
from sklearn.metrics import accuracy_score
print(accuracy_score(Y_test,pred))
```

```python
#unsing logistic regression
from sklearn.linear_model import LogisticRegression
model_lr=LogisticRegression()
model_lr.fit(X_train,Y_train) 
```

```python
pred1=model_lr.predict(X_test)
from sklearn.metrics import accuracy_score
print(accuracy_score(Y_test,pred1))
```

```python
#decision tree 
from sklearn.tree import DecisionTreeClassifier
model_dtc=DecisionTreeClassifier()
model_dtc.fit(X_train,Y_train)
```

```python
pred2=model_dtc.predict(X_test)
from sklearn.metrics import accuracy_score
print(accuracy_score(Y_test,pred2))
```

```python
#classification report
from sklearn.metrics import classification_report
print(classification_report(Y_test,pred2))
```

```python
#using elbow technique
from sklearn.cluster import KMeans

k_range = range(1, 10)
sse = []

for k in k_range:
    km = KMeans(n_clusters=k)  # Corrected 'n_clusters'
    km.fit(df[["petal_length", "petal_width"]])
    sse.append(km.inertia_)
```

```python
sse
```

```python
plt.xlabel("k_rng")
plt.ylabel("Sum of squared error")
plt.plot(k_range,sse)
```

```python
from sklearn.cluster import KMeans

km = KMeans(n_clusters=3, random_state=0)
y_predict = km.fit_predict(df[["petal_length", "sepal_length"]])

y_predict
```
